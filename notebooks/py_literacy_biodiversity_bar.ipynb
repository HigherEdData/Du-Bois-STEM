### 8. Adding Final Visual Layers and Titles With Your Own Name

To recreate the literacy bar chart exactly as it was originally drawn by Du Bois' lab, we have to add multiple layers of code. We've done so in the code cell below with layers that add the following: 
* Space for titles by setting the figure's width to 11 by height to 14.
* Change the background color
* Remove the number line at the bottom to declutter the image

Comments after #s in the code below explain which layers are added by new lines of code.

**Your Job**

To complete your recreation, fill in the blank in at the bottom of the new Layer 6 to add your own name to the subtitle: ```"Recreated by ___________",```. This will show that the graph was recreated by you!

<details> <summary><strong>Details on how the new code adds additional layers and chart elements.</strong></summary>

1. Add a 5th layer that declutters and improves legibility throuh the following steps:

- Flip y-axis order to go highest to lowest illiteracy with ```ax.invert_yaxis()``` command
- Set background color of plot with ```ax.set_facecolor()``` command
- Make country labels smaller and serif font with ```plt.yticks()``` command
- Remove x-axis ticks with ```ax.set_xticks([])```and remove x-axis tickslabels with ```ax.set_xlabel("")``` command
- Remove small tick marks from the y-axis with ```ax.tick_params()``` command
- Remove the chart border lines (spines) with ```for spine in ax.spines.values(): spine.set_visible(False)``` command

2. Add a 6th layer that creates a title space for the figure. This includes:

- Adjust position of the bar chart to make space for title with ```plt.subplots_adjust()``` command
- Place the title in the margin above the subplot with ```fig.text()``` command

3. Add a 7th layer to insert subtitle with another```fig.text()``` command

</details>

````python
# Layer 1: Import Packages and Load Data
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("d_literacy_country.csv") # Load the data from CSV

# Layer 2: Setup figure (overall chart area)
fig, ax = plt.subplots(figsize=(11, 14), facecolor="antiquewhite")

#  Layer 4: Apply Distinct Colors for Specific Variable
bar_colors =["red" if country == "Negroes, U.S.A."
    else "darkgreen" for country in df["country"]]

# Layer 3: Create a Horizontal Bar Chart

ax.barh(
    df["country"],  # y-axis df data frame source followed by var name in [] 
    df["illiteracy"],  # x-axis df data frame source followed by var name in [] 
    
# Addition to Layer 4:
    color= bar_colors, # apply the red and green colors from the bar_colors variable
    height=0.5 # set bar height
)

# New Layer 5: Mutliple formatting changes

ax.invert_yaxis() # flip y-axis order to go highest to lowest illiteracy

ax.set_facecolor("antiquewhite") # set background color of plot

plt.yticks(fontsize=10, fontname="serif") # make country labels smaller and serif font

ax.set_xticks([]) # Remove x-axis ticks
ax.set_xlabel("") # Remove x-axis tickslabels

ax.tick_params(axis="y", left=False) # Remove small tick marks from the y-axis

for spine in ax.spines.values(): spine.set_visible(False) # Remove the chart border lines (spines)

# Adjust position of the bar chart to make space for title
plt.subplots_adjust(left=0.25, right=0.85, top=.80, bottom=0.15)

# New Layer 6: Add Title with your name in subtitle
# Place the title in the margin ABOVE the subplot
fig.text(
    0.5, 0.95, # Position in middle at top
    "\n" # add a line of above the title so it's not to close to top edge
    "Illiteracy of the American Negroes compared with that of other nations.",
    fontsize=14, fontname="serif", weight="bold", ha="center"
)

fig.text(
    0.5, 0.85,  # # Position in middle just below title
    "Proportion d' illettrés parmi les Nègres Américains comparée à celle des autres nations."
    "\n\n" # add space between subtitle lines
    "Done by Atlanta University."
    "\n\n" # add space between subtitle lines
    "Recreated by _____________",
    fontsize=12, fontname="serif", ha="center"
)
````

## 10. Edit the Du Bois Graph Code to Plot the Biodiversity and Neighborhood Redlining Grade Data

After reading in the **d_biodiversity_redlining.csv** data, you can edit the exact graph code you used for the literacy code to graph the biodiversity and redlining data. Unlike the original biodiversity bar graph in the PNAS article, your code will graph the data as a horizontal bar graph.

````python
# Layer 1: Import Packages and Load Data
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("d_biodiversity_redlining.csv")

# Layer 2: Setup figure (overall chart area)
fig, ax = plt.subplots(figsize=(11, 14), facecolor="antiquewhite")

#  Layer 4: Apply Distinct Colors for Specific Variable
bar_colors = ["red" if grade == "D" else "darkgreen" for grade in df["grade"]]

# Layer 3: Create a Horizontal Bar Chart

# Create a horizontal bar chart in the ax subplot by telling the ax.barh function
# what data frame and variables to plot on the y-axis (bar name) and x-axis (bar length) 

ax.barh(
    df["____________"],  # (y-axis variable)
    df["____________"],  # (x-axis variable)
)

# Layer 5: Declutter and Improve Legibility of Figure

ax.invert_yaxis() # flip y-axis order to go highest to lowest illiteracy

ax.set_facecolor("antiquewhite") # set background color of plot

plt.yticks(fontsize=10, fontname="serif")  # make country labels smaller and serif font

ax.set_xticks([]) # Remove x-axis ticks
ax.set_xlabel("") # Remove x-axis tickslabels

ax.tick_params(axis="y", left=False)  # Remove small tick marks from y-axis

# Remove the chart border lines (spines)
for spine in ax.spines.values(): spine.set_visible(False)

# Layer 6: Adjust Figure and Create Title

# Adjust position of the bar chart to make space for title
plt.subplots_adjust(left=0.25, right=0.85, top=0.8, bottom=0.15)

# Place the title in the margin ABOVE the subplot
fig.text(
    0.5, 0.95, # Position in middle at top
    "\n" # add a line of above the title so it's not to close to top edge
    "San Diego Neighborhood Biodiversity Score (0-100) by Neighborhood Redlining Grade.",
    fontsize=12, fontname="serif", weight="bold", ha="center"
)

# Layer 7: Add Subtitle

# Fully center subtitle across figure
fig.text(
    0.5, 0.85, # X and Y position (X=0.5 ensures full centering)
# fill in the blank below to translate the title in the language of
# your choice
    "Adapted by __________ from Du Bois' graph of literacy in 1900 and from"
        "\n\n"
# Fill in the blank below to show the graph is adapted by you!
    "'Historical redlining is associated with disparities in wildlife biodiversity in four California cities' (2024)",
    fontsize=10, fontname="serif", ha="center"
)
````

### 11. Improve Accessibility by Adding X Axis Grid Lines and Removing the Use of Red and Green

Some of Du Bois' graphing choices might not make sense for graphs you want to make. 

For example, Du Bois also doesn't provide labels or grid lines to make it easy to understand what the range of biodiversity scores are by neighborhood redlining grade. We replaced the x tick and x label lines of code below with ```ax.xaxis.grid(zorder=0)``` to add grid lines.

In addition, red and green bars are difficult to differentiate for those with colorblindness.

To create a version of your graph in colorblind friendly colors, edit Layer 4 by :
1. Filling in the blank to set color for the D grade bar to **orange** or another color that expresses your creativity.
2. change the color for the other bars to **blue** or another color that expresses your creativity.
3. Fill in the blank again to add your name.

````python
# Layer 1: Import Packages and Load Data
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("d_biodiversity_redlining.csv")

# Layer 2: Setup figure (overall chart area)
fig, ax = plt.subplots(figsize=(11, 14), facecolor="antiquewhite")

# Layer 4: Apply Distinct Colors for Specific Variable
bar_colors = ["__________" if grade == "D" else "__________" for grade in df["grade"]]

# Layer 3: Create a Horizontal Bar Chart

# Create a horizontal bar chart in the ax subplot by telling the ax.barh function
# what data frame and variables to plot on the y-axis (bar name) and x-axis (bar length) 

ax.barh(
    df["grade"],  # (y-axis)
    df["biodiversity"],  # (x-axis)

# Addition to Layer 4:
    color=bar_colors,  # apply red and green colors to the bars with the color bar color variable
    # fill in the blank below to set the bar height
    height=0.5
)

# Layer 5: Declutter and Improve Legibility of Figure

ax.invert_yaxis() # flip y-axis order to go highest to lowest illiteracy

ax.set_facecolor("antiquewhite") # set background color of plot

plt.yticks(fontsize=10, fontname="serif")  # make country labels smaller and serif font

ax.xaxis.grid(zorder=0)  # Add xaxis grid variable and set zorder to 0 so lines appear behind bars

ax.tick_params(axis="y", left=False)  # Remove small tick marks from y-axis

# Remove the chart border lines (spines)
for spine in ax.spines.values(): spine.set_visible(False)

# Layer 6: Adjust Figure and Create Title

# Adjust position of the bar chart to make space for title
plt.subplots_adjust(left=0.25, right=0.85, top=0.8, bottom=0.15)

# Place the title in the margin ABOVE the subplot
fig.text(
    0.5, 0.95, # Position in middle at top
    "\n" # add a line of above the title so it's not to close to top edge
    "San Diego Neighborhood Biodiversity Score (0-100) by Neighborhood Redlining Grade.",
    fontsize=12, fontname="serif", weight="bold", ha="center"
)

# Layer 7: Add Subtitle

# Fully center subtitle across figure
fig.text(
    0.5, 0.85, # X and Y position (X=0.5 ensures full centering)
# fill in the blank below to translate the title in the language of your choice
    "Adapted by __________ from Du Bois' graph of literacy in 1900 and from"
        "\n\n"
# Fill in the blank below to show the graph is adapted by you!
    "'Historical redlining is associated with disparities in  wildlife biodiversity in four California cities' (2024)",
    fontsize=10, fontname="serif", ha="center"
)
````
