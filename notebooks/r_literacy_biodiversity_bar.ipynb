### 8. Adding Final Visual Layers and Titles With Your Own Name

To recreate the literacy bar chart exactly as it was originally drawn by Du Bois' lab, we have to add multiple layers of code. We've done so in the code cell below with layers that add the following: 
* Space for titles by setting the figure's width to 11 by height to 14.
* Change the background color
* Remove the number line at the bottom to declutter the image

Comments after #s in the code below explain which layers are added by new lines of code.

**To complete your recreation,** fill in the blank in the code below for a new Layer 3 to put subtitle text in the second **labs** function: ```"Recreated by ___________\n\n",``` to add your own name. This will show that the graph was recreated by you!

<details> <summary><strong>Details on how the new code adds additional layers and chart elements.</strong></summary>

Key new lines of code do the following:

1. Source the Du Bois theme from a file at the STEM + Du Bois site: ```source("https://stemdubois.org/theme_dubois.R)``` to direct the section to a source for theming the figure

2. Add a width specification,```(width = .5)```, to the geom_col() bar graph command.

3. Use plus signs ```+``` after the ```geom_col()``` line of code to tell R you are adding additional lines of code

4. Specify ```theme_dubois()``` on a new line after a ```+``` sign. This will apply several Du Bois style options like an "antique white" background color and minimal text in labeling axes. You can see all of these R theme options in the "theme_dubois.R" theme file here: https://github.com/HigherEdData/Du-Bois-STEM/blob/main/theme_dubois.R

5. Add ```scale_fill_manual``` function of ggplot below to set the colors of the bars as red or darkgreen based on whether it is true or false that a given country category is "Negroes, U.S.A."

6. Include the blank of the ```element_text``` function with the fontname serif to change the Du Bois theme's font for the graph to a serif font.

</details>

````r
# Layer 1: Load Data, Library, and Source Theme
df <- read.csv("d_literacy_country.csv")
library(ggplot2)
options(repr.plot.width=22/2, repr.plot.height=28/2)
source("https://stemdubois.org/theme_dubois.R")

# Layer 2: Set the ggplot call with added bar colors
ggplot(df, aes(
    x = illiteracy,
    y = reorder(country, illiteracy),
    fill = country == "Negroes, U.S.A." # this is the fill statement
)) +

# New Layer 3: Formatting changes then and set your name in the title
scale_fill_manual(values = c("TRUE" = "red", "FALSE" = "darkgreen")) +
geom_col(width = .5) +
theme_dubois() +
theme(text = element_text(family = "serif"), axis.text.y = element_text(size = 16)
) +

labs(
    title = "\nIlliteracy of the American Negroes compared with that of other nations.\n",
    subtitle = "Proportion d' illettrés parmi les Nègres Americains comparée à celle des autres nations.\n\n
    Done by Atlanta University.\n\n
    Recreated by ________\n\n"
    )
````

### 11. Improve Accessibility by Adding X Axis Grid Lines and Removing the Use of Red and Green

Some of Du Bois' graphing choices might not make sense for graphs you want to make. 

For example, Du Bois doesn't provide labels or grid lines to make it easy to understand what the range of biodiversity scores are by neighborhood redlining grade. We added lines of code in Layer 5 below to add grid lines to the graph.

In addition, red and green bars are difficult to differentiate for those with colorblindness.

**To create a version of your graph in colorblind friendly colors:**
1. Fill in the blank to set the bar colors to orange and blue in this line of code: ```scale_fill_manual(values = c("TRUE" = "_____", "FALSE" = "______")) +```.
2. Fill in the blank to add your name again.

If you want to customize the chart further to add your own style twist, try copying and pasting the code into an AI assistant and ask it how to change the code for your customization of choice.

````r
# Layer 1: Load Data, Library, and Source Theme
df<- read.csv("d_biodiversity_redlining.csv")
library(ggplot2)
options(repr.plot.width=22/2, repr.plot.height=28/2)
source("https://stemdubois.org/theme_dubois.R")

# Layer 2: Set the ggplot call with added bar colors
ggplot(df, aes(
    x = biodiversity,
    y = reorder(grade, biodiversity),
    fill = grade == "D" # this is the fill statement
)) +

# Layer 3: Formatting changes and set your name in the title
scale_fill_manual(values = c("TRUE" = "________", "FALSE" = "________")) +
geom_col(width = .5) +
theme_dubois() +
theme(text = element_text(family = "serif"), axis.text.y = element_text(size = 16)
) +

labs(
    title = "\nSan Diego Neighborhood Biodiversity Score (0-100) by Neighborhood Redlining Grade.\n",
    subtitle = "Adapted by ________from Du Bois' graph of literacy in 1900 and from\n
    'Historical redlining is associated with disparities in wildlife biodiversity in four California cities' (2024)\n\n"
    ) +

# Layer 4: Add Grid Lines with Labels
scale_x_continuous(
    breaks = seq(0, 60, by = 10),  # Set tick positions every 10 units
    ) +
theme(
    axis.text.x = element_text(size = 16),
    panel.grid.major.x = element_line(color = "lightgray")
    )
````
